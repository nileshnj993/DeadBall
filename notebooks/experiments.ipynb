Find the unique set piece labels in the dataset so we can create an ENUM to refer to them easily.

```python
from statsbombpy import sb

competitions = sb.competitions()
matches = sb.matches(11, 90)
unique_play_patterns = set()
for index, row in matches.iterrows():
    events = sb.events(row['match_id'])
    unique_play_patterns.update(events['play_pattern'].unique())
print(unique_play_patterns)
```
